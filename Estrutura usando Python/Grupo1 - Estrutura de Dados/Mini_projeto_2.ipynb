````python
def idade_min_max_aprovados(dados):
  aprovados = [registro for registro in dados if registro['aprovado'] == 'true']
  idades = [int(registro['idade']) for registro in aprovados]
  idade_min = min(idades)
  idade_max = max(idades)
  return (idade_min, idade_max)
````

````python
def filtrar_duracao_maior_que(dados, meses):
  return [registro for registro in dados if int(registro['duracao_meses']) > meses]
````

````python
def filtrar_moradia_aluguel(dados, tipo):
    return [item for item in dados if item.get("moradia") == tipo]
````

````python
def filtrar_5_maiores_rendaAnual_list(dados):
  salarios_ordenados = sorted(dados, key=lambda x: int(x['salario_anual']), reverse=True)
  return salarios_ordenados[:5]
````

````python


def filtrar_5_maiores_rendaAnual_deque(dados):
  from collections import deque
  salarios_ordenados = deque(sorted(dados, key=lambda x: int(x['salario_anual']), reverse=True))
  return list(salarios_ordenados)[:5]
````

````python
def filtrar_5_maiores_rendaAnual(dados):
  import timeit

  return (timeit.timeit('filtrar_5_maiores_rendaAnual_list(dados)', number=1000, globals=globals()), timeit.timeit('filtrar_5_maiores_rendaAnual_deque(dados)', number=1000, globals=globals()))
````

````python
import json
import unicodedata
from defs import (
    filtrar_moradia_aluguel,
    filtrar_duracao_maior_que,
    idade_min_max_aprovados,
    filtrar_5_primeiros_aprovados,
    filtrar_5_últimos_original,
    filtrar_5_maiores_rendaAnual,
)

with open("aprovacoes_emprestimo.json", "r", encoding="utf-8") as f:
    dados = json.load(f)


def _normalize(s: str) -> str:
    if not isinstance(s, str):
        return ""
    s = s.lower().strip()
    s = unicodedata.normalize("NFKD", s)
    return "".join(ch for ch in s if not unicodedata.combining(ch))



def app():
    while True:
        # interface de usuário
        print("\nEscolha uma opção:")
        print("1 - Filtrar por moradia")
        print("2 - Filtrar por duração maior que X meses")
        print("3 - Idade mínima e máxima dos aprovados")
        print("4 - Filtrar os 5 primeiros aprovados")
        print("5 - Filtrar os 5 últimos registros originais")
        print("6 - Filtrar os 5 maiores salários anuais")
        print("0 - Sair")

        escolha = input("Digite o número da opção desejada: ").strip()

        if escolha == "1":
            tipo = input("Digite o tipo de moradia (própria/alugada/aluguel): ").strip()


            def _match(item):
                return _normalize(item.get("moradia", "")) == _normalize(tipo)

            resultado = [it for it in dados if _match(it)]


            print(f"Resultados para moradia '{tipo}': {resultado}")

        elif escolha == "2":
            meses = int(input("Digite a duração mínima em meses: ").strip())
            resultado = filtrar_duracao_maior_que(dados, meses)
            print(f"Resultados para duração maior que {meses} meses: {resultado}")

        elif escolha == "3":
            idade_min, idade_max = idade_min_max_aprovados(dados)
            print(f"Idade mínima dos aprovados: {idade_min}, Idade máxima dos aprovados: {idade_max}")

        elif escolha == "4":
            resultado = filtrar_5_primeiros_aprovados(dados)
            print(f"5 primeiros aprovados: {resultado}")

        elif escolha == "5":
            resultado = filtrar_5_últimos_original(dados)
            print(f"5 últimos registros originais: {resultado}")

        elif escolha == "6":
            timeList, timeDeque = filtrar_5_maiores_rendaAnual(dados)
            print(f"Tempo usando list: {timeList}, Tempo usando deque: {timeDeque}")

        elif escolha == "0":
            print("Saindo...")
            break

        else:
            print("Opção inválida. Tente novamente.")


if __name__ == "__main__":
    app()
````
