## Importando e criando um nó

```python
from binarytree import Node

# Cria uma árvore manual
raiz = Node(20)
raiz.left = Node(10)
raiz.right = Node(30)
raiz.left.left = Node(5)
raiz.left.right = Node(15)
raiz.right.right = Node(35)
```

```python
print(raiz)
```

```python
# via graphviz
raiz.graphviz()
```

## Gerando uma árvore binária aleatória

```python
from binarytree import tree

arv = tree(height=9,is_perfect=False)
arv.graphviz()
```

## Gerando uma árvore binária de busca (ABB) aleatória

```python
from binarytree import bst

ABB = bst(height = 3)
ABB.graphviz()
```

## Convertendo lista em árvore binária

```python
from binarytree import build

'''
Método build:
>> Nó Pai está no índice i,
>>> Filho à esquerda: 2i + 1
>>> Filho à direita em 2i + 2
>>> 'None' indica ausência de um nó
'''
nodes = [1,2,3,None,4,5]
arv = build(nodes)
arv.graphviz()
```

## Métodos úteis da biblioteca

```python
print("Maior nó:", raiz.max_node_value )
print("Menor nó:",raiz.min_node_value )
print("Número de folhas:", raiz.leaf_count )
print("É uma árvore binária de busca (ABB)?", raiz.is_bst )
print("Lista dos valores:", raiz.values )
```

### **Exercício**

* Gere uma BST utilizado a função recursiva a seguir e a função Node() da biblioteca Binarytree.
* Valide se a BST foi gerada com sucesso a partir do método `is_bst`.

```python
# Função Inserir Recursiva
def inserir(raiz, valor):
    if raiz is None:
        return Node(valor)

    if valor < raiz.value:
        raiz.left = inserir(raiz.left, valor)
    elif valor > raiz.value:
        raiz.right = inserir(raiz.right, valor)
    # Caso valor == raiz.value, não faz nada (ignora duplicata)
    return raiz
```
