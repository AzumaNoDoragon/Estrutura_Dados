### Exercício 1 – Calculadora de IMC

Faça um programa que:

1. Leia do usuário o **peso** (kg) e a **altura** (m).
2. Calcule o **IMC = peso / (altura²)**.
3. Exiba a classificação:

   * Menor que 18.5 → "Abaixo do peso"
   * Entre 18.5 e 24.9 → "Peso normal"
   * Entre 25.0 e 29.9 → "Sobrepeso"
   * Acima de 30.0 → "Obesidade"

````python
peso = float(input("Informe o peso (Kg): "))
altura = float(input("Informe a altura (m): "))

imc = peso / (altura**2)
# print(imc)

if imc < 18.5:
  print("Abaix do peso")
elif imc >= 18.5 and imc <= 24.9:
  print("Peso normal")
elif imc >= 25.0 and imc <= 29.9:
  print("Sobrepeso")
else:
  print("Obesidade")
````

### Exercício 2 – Verificador de Número

Faça um programa que leia um número inteiro e:

* Verifique se ele é **positivo, negativo ou zero**.
* Informe também se é **par ou ímpar**.


````python
n = int(input("Digite um número: "))

# Verifica se é positivo, negativo ou zero
if n > 0:
  print("Positivo!"])
elif n < 0:
  print("Negativo!")
else:
  print("Zero!")

# Verifica se é pa ou ímpar
if n % 2 == 0:
  print("O valor é par!")
else:
  print("O valor é ímpar!")
````

### Exercício 3 – Contagem Regressiva

Faça um programa que leia um número inteiro positivo **N** e exiba uma contagem regressiva até 0.
Ao final, mostre a mensagem `"Fim da contagem!"`.




````python
n = int(input("Digite um valor: "))
for i in range(n,-1,-1):
  print(i)
print("Fim da contagem!")
````

### Exercício 4 – Fatorial de um Número

Escreva um programa que leia um número inteiro positivo e calcule o seu **fatorial** usando um laço de repetição.

> Exemplo:
> `5! = 5 × 4 × 3 × 2 × 1 = 120`



````python
n = int(input("Digite um valor: "))
fat = 1
for i in range(n,0,-1):
 fat *= i
print(f"Fatorial de {n} == {fat}")
````
